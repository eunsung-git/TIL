```python
# matplotlib : matplotlib.org -> 코드 복사하여 사용 가능

import matplotlib.pyplot as plt

data = {'apples': 10, 'oranges': 15, 'lemons': 5, 'limes': 20}
names = list(data.keys())
values = list(data.values())

fig, axs = plt.subplots(1, 3, figsize=(9, 3), sharey=True)
axs[0].bar(names, values)
axs[1].scatter(names, values)
axs[2].plot(names, values)
fig.suptitle('Categorical Plotting')
```

```python
## xticks / yticks   ->   plot의 축 위치 표시 지점

import numpy as np

#np.np.linspace(a,b,n)    -> a부터 b까지 n개로 분할
x=np.linspace(-np.pi,np.pi,256)
c=np.cos(x)

plt.plot(x,c)
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi])
plt.yticks([-1,0,1])
plt.show

```

```python
# plot 여러 개 출력
t=np.arange(0,5,0.2)
plt.plot(t,t,'r--',t,0.5*t**2,'bs:')
plt.show()
```

```python
## 범례 추가
x=np.linspace(-np.pi,np.pi,256)
c,s=np.cos(x),np.sin(x)

plt.plot(x,c,ls='--',label='cos')
plt.plot(x,s,ls=':',label='sin',c='red')
plt.legend(loc=0)
plt.show()
```

```python
#figure  -> paper
# axes   -> plot
# axis   -> 축

np.random.seed(0)
f1=plt.figure(figsize=(10,2))    # figure size 설정
plt.plot(np.random.randn(100))
plt.show()
```

```python
#subplot(a,b,n)   #subplot(a행,b열,a중에 n번째 위치)
x1=np.linspace(0.0,5.0)
x2=np.linspace(0.0,2.0)
y1=np.cos(2*np.pi*x1)*np.exp(-x1)
y2=np.cos(2*np.pi*x2)

ax1=plt.subplot(2,1,1)
plt.plot(x1,y1,'yo-')
plt.title('table1')

ax2=plt.subplot(2,1,2)
plt.plot(x2,y2,'r.-')
plt.title('table2')

plt.tight_layout()

plt.show()
```

```python
tips=sns.load_dataset('tips')
tips.columns
```

```python
hist=plt.figure()
axes1=hist.add_subplot(1,1,1)
axes1.hist(tips['total_bill'],bins=6)

scplot=plt.figure()
axes1=scplot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'],tips['tip'])

boxplot=plt.figure()
axes1=boxplot.add_subplot(1,1,1)
axes1.boxplot([tips[tips['sex']=='Female']['tip'], tips[tips['sex']=='Male']['tip']],labels=['Female','Male'])
```

```python
tips['sex_color']=tips['sex'].map({'Female':0, 'Male':1})
#or
def enc(s):
    if s=='Female':
        return 0
    else:
        return 1
tips['sex_color']=tips['sex'].apply(enc)

scplot=plt.figure()
axes1=scplot.add_subplot(1,1,1)
# alpha : 투명도
axes1.scatter(x=tips['total_bill'],y=tips['tip'],s=tips['size']*10,c=tips['sex_color'],alpha=0.8)
```
