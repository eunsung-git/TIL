```python
df2=pd.DataFrame({'c1':[1,2,3,4,5],
            'c2':[6,7,8,9,10]})
df2.ix[[1,3],['c2']]=np.nan
df2
```

```python
df2['c2p']=np.where(pd.notnull(df2['c2'])==True,df2['c2'],df2['c1'])
df2
```

```python
for i in df2.index:
    if pd.notnull(df2.ix[i,'c2'])==True:
        df2.ix[i,'c2p']=df2.ix[i,'c2']
    else:
        df2.ix[i,'c2p']=df2.ix[i,'c1']
df2
```
